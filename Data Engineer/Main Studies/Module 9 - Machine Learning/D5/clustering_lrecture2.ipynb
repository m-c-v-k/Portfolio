```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn import datasets
%matplotlib inline
```

```python
# Import data
iris = datasets.load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
targets = iris.target_names
df
```

```python
# Define our X, the data that we are clustering
X = iris.data
```

```python
plt.scatter(X[:,0], X[:,1], c=df['target'])
```

```python
# Plot the values in 3D
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['sepal length (cm)'], df['petal length (cm)'], df['petal width (cm)'], c=df['target'])

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Petal Length')
ax.set_zlabel('Petal Width')
```

```python
# Define our empty model
from sklearn.cluster import KMeans

km = KMeans(3)
```

```python
# Fit/train the model with data
clusters = km.fit_predict(X)
```

```python
plt.scatter(X[:,2], X[:,3], c=km.labels_)
```

```python
# Plot the values in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(
    km.cluster_centers_[:,0],
    km.cluster_centers_[:,2],
    km.cluster_centers_[:,3],
    s=250,
    c='red')
scatter = ax.scatter(
    df['sepal length (cm)'],
    df['petal length (cm)'],
    df['petal width (cm)'],
    c=df['target'])

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Petal Length')
ax.set_zlabel('Petal Width')
```
