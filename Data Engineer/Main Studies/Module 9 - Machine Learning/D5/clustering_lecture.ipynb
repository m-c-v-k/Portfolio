```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
# Create data to be clustered
X = np.array([[1, 1], [2, 1], [2, 2], [1, 2], [3, 3], [3, 4], [4, 4], [4, 3], [5, 5], [5, 6], [6, 6], [6, 5]])
# Examine our data
plt.scatter(X[:,0], X[:,1])
```

```python
# Create empty model
km = KMeans(3, random_state=7)
```

```python
# Train the model
km.fit_predict(X)
```

```python
# Evaluate the model
print(km.labels_)
print(X)
```

```python
# Evaluate the result in a plot
# The labels from our model corresponds to our samples
# and can be used to color the plot

plt.scatter(X[:,0], X[:,1], c=km.labels_)
```
