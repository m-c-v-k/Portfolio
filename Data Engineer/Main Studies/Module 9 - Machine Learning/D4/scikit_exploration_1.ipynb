### 9.21 Decision tree exercise

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
# Get data
iris = load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

- Name features in tree
- Play around with parameters in plot_tree
- Show class names for all results in tree

```python
# Classify and fit data
attribute_names = [
    'sepal length in cm',
    'sepal width in cm',
    'petal length in cm',
    'petal width in cm'
    ]
class_names = iris.target_names

clf  = DecisionTreeClassifier(max_leaf_nodes=4, random_state=0)
clf.fit(X_train, y_train)
tree.plot_tree(clf, rounded=True, fontsize=7, filled=True, feature_names=attribute_names, impurity=True, proportion=True, class_names=class_names)
plt.show()
```
